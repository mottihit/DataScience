```python
#Single cards - football 
link = "https://goldin.co/buy/?show_only=Sold%20Items&page=1&number_of_lots=96&Item_Type=Single%2520Cards&Sub_Category=Football"
driver =webdriver.Chrome()
driver.get(link)
driver.implicitly_wait(10)
cards = driver.find_elements(By.CSS_SELECTOR,".css-175oi2r.r-1udh08x.r-y46g1k")

```
